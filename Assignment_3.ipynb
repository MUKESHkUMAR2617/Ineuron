```
#Task 1
'''
Q:1 Write a Python Program to implement your own myreduce() function which works exactly
like Python's built-in function reduce()
'''
def myreduce(function,lst):
  b=lst[0]
  for i in range(1,len(lst)):
    b= function(b,lst[i])
  return b

def mul(x,y):
  return x*y

def sub(x,y):
  return x-y

def sum(x,y):
  return x+y

a=[9,8,7,6,5,4,3,2,1] 

print('multiply of list a is:',myreduce(mul,a))

print('subtract of list a is:',myreduce(sub,a))

print('sum of list a is:',myreduce(sum,a))

```

```
#Task 2
'''
Q.no 2: 1.2 Write a Python program to implement your own myfilter() function which works exactly
like Python's built-in function filter()
'''
def myfilter(function,lst):
  a=[]
  b=[]

  for i in lst:
    a.append(function(i))

  for i in range(0,len(a)):
    if a[i]== False:
      continue

    else:
      b.append(a[i])

  return b

def even_num(i):
    
    if i%2==0:
        return i
    
    else:
        return False

def odd_num(i):
    if i%2 == 1:
        return i
    else:
        return False        
print(myfilter(even_num,[1,2,3,4,5,6,7,8,9]))
print(myfilter(odd_num,[9,8,7,6,5,4,3,2,1]))
```
